```python
import os
import pytest
import nest_asyncio
nest_asyncio.apply()

HOSTNAME = os.environ.get("REDPITAYA_HOSTNAME")
PASSWORD = os.environ.get("REDPITAYA_PASSWORD")

# 👇 Skip instead of hard-failing if not configured
if HOSTNAME is None:
    pytest.skip("REDPITAYA_HOSTNAME not set", allow_module_level=True)

if PASSWORD is None:
    PASSWORD = 'root'

from pyrpl import Pyrpl

p = Pyrpl(
    hostname=HOSTNAME,
    password=PASSWORD,
    config='',  # no config file
)

p.rp.scope.setup(duration=0.0001, trace_average=1)
curve = p.rp.scope.single()

assert curve.shape == (2, 2**14)
```
